```python
import numpy as np 
from model import load_dataset, SynapticDownscalingModel, train_model, plot_accuracies, test_model
import matplotlib.pyplot as plt
```

```python
train_dataset = load_dataset(train=True)
test_dataset = load_dataset(train=False)
num_epochs = 10
```

```python
base_model = SynapticDownscalingModel()
```

```python
train_accuracies, test_accuracies, train_losses, per_task_test_accuracies = train_model(base_model, train_dataset, test_dataset, epochs_per_task=num_epochs, nrem_replay=False)
```

```python
task_epochs = num_epochs
epochs = list(range(1, task_epochs*5+1))
plot_accuracies(train_accuracies, test_accuracies, epochs, task_epochs)
```

```python
test_model(base_model, test_dataset)
print(len(per_task_test_accuracies))
print(len(per_task_test_accuracies[0]))
```

```python
downscaling_model_p25 = SynapticDownscalingModel()
train_accuracies_p25, test_accuracies_p25, train_losses_p25, per_task_test_accuracies_p25 = train_model(downscaling_model_p25, train_dataset, test_dataset, nrem_replay=False, synaptic_downscaling=True, p=0.25, epochs_per_task=num_epochs)
plot_accuracies(train_accuracies_p25, test_accuracies_p25, epochs, task_epochs)
```

```python
downscaling_model_p50 = SynapticDownscalingModel()
train_accuracies_p50, test_accuracies_p50, train_losses_p50, per_task_test_accuracies_p50 = train_model(downscaling_model_p50, train_dataset, test_dataset, nrem_replay=False, synaptic_downscaling=True, p=0.5, epochs_per_task=num_epochs)
plot_accuracies(train_accuracies_p50, test_accuracies_p50, epochs, task_epochs)
```

```python
downscaling_model_p75 = SynapticDownscalingModel()
train_accuracies_p75, test_accuracies_p75, train_losses_p75, per_task_test_accuracies_p75 = train_model(downscaling_model_p75, train_dataset, test_dataset, nrem_replay=False, synaptic_downscaling=True, p=0.75, epochs_per_task=num_epochs)
plot_accuracies(train_accuracies_p75, test_accuracies_p75, epochs, task_epochs)
```

```python
num_tasks = 5
colors = plt.cm.get_cmap('tab10', num_tasks)
task_labels = [f"Task {t}" for t in range(num_tasks)]

#fig, axes = plt.subplots(1, 4, sharey=True)

model_tests = [per_task_test_accuracies,per_task_test_accuracies_p25, per_task_test_accuracies_p50, per_task_test_accuracies_p75]
labels = ['No Downscaling', 'p=0.25', 'p=0.5', 'p=0.75']
for i in range(len(model_tests)):
    task_test_accuracies = [[], [], [], [], []]
    for j in range(num_tasks):
        for k in range(len(model_tests[i])):
            task_test_accuracies[j].append(model_tests[i][k][j]*0.2)
    task_test_accuracies = np.array(task_test_accuracies)
    task_test_accuracies[1][task_epochs:]+=task_test_accuracies[0][task_epochs:]
    task_test_accuracies[2][task_epochs*2:]+=task_test_accuracies[1][task_epochs*2:]
    task_test_accuracies[3][task_epochs*3:]+=task_test_accuracies[2][task_epochs*3:]
    task_test_accuracies[4][task_epochs*4:]+=task_test_accuracies[3][task_epochs*4:]
    #print(task_test_accuracies)
    plt.figure()
    for t in range(num_tasks):
        plt.plot(range(t*task_epochs+1, len(task_test_accuracies[t])+1), task_test_accuracies[t][t*task_epochs:], label=task_labels[t])
        plt.xlabel('Epochs')
        plt.ylim(bottom=0, top=0.8)
        plt.ylabel('Test Accuracy')
        plt.title(labels[i] + ' - Per Task Test Accuracies')
        plt.legend()
    plt.show()
```
